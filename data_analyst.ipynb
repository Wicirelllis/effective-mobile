### Задание 1.1: Фермер

На ферме содержатся шесть разных видов животных, и каждый раз, когда фермер заходит в сарай, он видит одно случайное животное. За день фермер заходит в сарай 6 раз.

Каково математическое ожидание количества разных видов животных, которые фермер увидит за день?

Разобъем нужное событие на шесть индикаторов - $A_i$.
$A_i$ равен 1, если за все шесть посещений фермер видел вид $i$ хотя бы один раз и 0 иначе.
Матожидание такого индикатора - $E(A_i) = P(A_i = 1) = 1 * (1 - (5/6)^6)$.
Матожидание из задачи, из-за линейности матожидания, $6 * E(A_i) = 6 * (1 - (5/6)^6) \approx 3.99$

### Ответ: 3,99

### Задание 1.2: Кулинарное соревнование

В конкурсе участвуют 80 шеф-поваров с уникальными уровнями мастерства. В первом этапе судьи случайным образом распределяют их по парам (в любом состязании двух шефов выигрывает тот, у кого выше уровень мастерства). На втором этапе шефы снова случайно образуют пары для финального раунда (пары могут повториться). Победную награду получают те, кто выиграл в обоих этапах.

Каково математическое ожидание числа победителей?

Пронумеруем поваров по возрастанию мастерства от 1 до 80.
Тогда шеф $k$ побеждает всех с меньшими номерами и проигрывает всем с большими.
Таким образомм, вероятность победить в раунде $(k-1)/79$.
Раундов два, они независимы, поэтому итоговая вероятность, что шеф $k$ победит - $(k-1)^2 / 79^2$.
Матожидание числа победителей - сумма $\sum_{k=1}^{80} (k-1)^2 / 79^2 = \sum_{k=0}^{79} k^2 / 79^2 = 1/79^2 * (79 * (79+1) * (2*79+1)) / 6 \approx 26.8$

### Ответ: 26,8

### Задание 1.3: Одинокая дорога

На пустынном шоссе вероятность появления автомобиля за 30-минутный период составляет 0.95.

Какова вероятность его появления за 10 минут? А за 27 минут?

Будем считать, что появление автомобиля описывается распределением Пуассона.
Т.е. вероятность того, что за $t$ минут проедет $k$ автомобилей равна $(\lambda t)^k e^{-\lambda t} / k!$
Из условия получаем $e^{-30\lambda} = 0.05 \implies \lambda = \ln{20} / 30$

Вероятность появления автомобиля за 10 минут:
$1 - e^{-10 * ln20/30} = 1 - 20^{-1/3} \approx 0.632$

Вероятность появления автомобиля за 27 минут:
$1 - e^{-27 * ln20/30} = 1 - 20^{-27/30} \approx 0.933$

### Ответ: 63,2; 93,3

### Задание 2.1: Изоморфизмы

Реализовать функцию (или тело функции), которая проверяет на изоморфность два
слова. Пояснение: строки s и t называются изоморфными, если все вхождения каждого символа строки s можно последовательно заменить другим символом и получить строку t. Порядок символов при этом должен сохраняться, а замена — быть
уникальной. Так, два разных символа строки s нельзя заменить одним и тем же
символом из строки t, а вот одинаковые символы в строке s должны заменяться одним и тем же символом.

````python
def is_isomorphic(s: str, t: str) -> bool:
  if len(s) != len(t):
    return False
  mapping = {}
  for a, b in zip(s, t):
    if a not in mapping:
      mapping[a] = b
    else:
      if mapping[a] != b:
        return False
  return True
````

````python
tests = [
    ['paper', 'title'],
    ['', ''],
    ['a', 'z'],
    ['aa', 'wb'],
    ['aba', 'cdc'],
    ['aa', 'bbc']
]

for s, t in tests:
  print(is_isomorphic(s, t))
````

Сложность по времени - линейная. Плюс, решение работает за один проход по каждой из строк. Лучше нельзя, потому что, как минимум, нужно прочитать каждую строку, а это уже линия. Ранний выход, если понятно, что ответ "нет" - бонус, но на сложность в худшем случае не влияет.

Затраты по памяти - константные (на словарь `mapping`, он ограничен размером алфавита)

### Задание 2.2: Натуральная последовательность

Реализовать функцию (или тело функции), которая находит единственное отсутствующее число из последовательности натуральных чисел 1,2,…,n.

````python
def missing_number(nums: list[int]) -> int:
  n = len(nums) + 1
  return n * (n + 1) // 2 - sum(nums)
````

Аналогично предыдущей задаче, сложность по времени - линейная, за один проход.
По памяти - константная. Значительно улучший невозможно.

### Задание 2.3: Факторизация

Реализовать функцию (или тело функции), которая при введении натурального числа n разбивает его на простые множители (представить его в виде простых чисел).

````python
def is_prime(n: int, primes: list[int]) -> bool:
  for i in primes:
    if n % i == 0:
      return False
  return True

def prime_factors(n: int) -> list[int]:
  if n == 1:
    return [1]
  primes = [2]
  result = []
  while n > 1:
    d = primes[-1]
    if n % d == 0:
      result.append(d)
      n = n // d
    else:
      while not is_prime(d, primes):
        d += 1
      primes.append(d)
  return result
````

Ищем делители среди простых чисел, от меньших к большим.
Лист простых строим в процессе.
Проверяем на простоту делением на все меньшие простые - не делится ни на одно? - простое.

Решение вполне оптимально, построение списка простых можно было бы сделать чуть лучше, но ценой заметно более сложного кода.
Сложность по памяти - $O(n/\ln n)$ на хранение списка простых.
Решение работает в пределах секунды на числах до $10^{10}$.

общем рейтинге.

```sql

SET rank = OtherTable.rank
```

покупок меньше, чем на 5000 рублей.

```sql
SELECT a.client_id
FROM account a
LEFT JOIN transaction t
    ON a.id = t.account_id
```

### Задание 3: Ручной счёт корреляции
Рассчитайте линейную корреляцию Пирсона, на основе данных.
Какой вывод можно сделать на основе полученного результата? Можно ли утверждать, что существует причинно-следственная связь между количеством чашек кофе, выпитых студентами в течение экзаменационного дня, и их итоговым баллом за экзамен?

````python
x = [1, 1, 2, 2, 2, 2, 3, 3, 3, 4]
y = [85, 88, 79, 81, 84, 65, 67, 58, 76, 49]

x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)
x_ = [i - x_mean for i in x]
y_ = [i - y_mean for i in y]
corr = sum([i * j for i, j in zip(x_, y_)]) / (sum(i ** 2 for i in x_) * sum(j ** 2 for j in y_)) ** 0.5
corr
````

Линейная кокрреляция примерно равна $-0,85$ - сильная отрицательная корреляция, больше чашек - ниже балл.
Но говорить о причинно-следственной связи нельзя, потому что "Correlation does not imply causation"
